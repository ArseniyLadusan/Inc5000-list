```python
# Years on List distribution
sns.countplot(x='yrs_on_list', data=inc5000)
plt.show()
```

```python
# Industry Distribution
sns.countplot(y='industry', data=inc5000)
plt.show()
```

```python
# year by year of industries listed
inca = sns.FacetGrid(inc5000, col="industry",col_wrap=5,)
inca= inca.map(plt.hist, "year",)
inca.set_titles("{col_name}")
plt.show()
```

```python
# industry facetgrid with yrs_on_list on x-axis
incb = sns.FacetGrid(inc5000, col="industry",col_wrap=5,)
incb= incb.map(plt.hist, "yrs_on_list",)
incb.set_titles("{col_name}")
plt.show()
```

```python
ax= sns.regplot(x=inc5000["yrs_on_list"], y=inc5000["growth"], fit_reg=False)
ax.set(xlabel='Years On List', ylabel='Growth %')
plt.show()
```

```python
inc5000['yrs_on_list'].mean()
```

```python
ax= sns.regplot(x=inc5000["revenue"], y=inc5000["growth"], fit_reg=False)
ax.set(xlabel='Revenue', ylabel='Growth %')
plt.show()
```

```python
# Scatterplot Facetgrid with years on list x-axis
incb = sns.FacetGrid(inc5000, col="industry",col_wrap=5,)
incb= incb.map(plt.scatter,"yrs_on_list","growth")
incb.set_titles("{col_name}")
plt.show()
```

```python
# Scatterplot Facetgrid with revenue on x-axis
incb = sns.FacetGrid(inc5000, col="industry",col_wrap=5,)
incb= incb.map(plt.scatter,"revenue","growth")
incb.set_titles("{col_name}")
plt.show()
```

```python
# Filter the outliers
rev = inc5000['revenue'] <= 1000000000
gro = inc5000['growth'] <= 100000

inc5000 = inc5000[rev & gro]
```

```python
# Plotting the Growth rate and Revenue
ax= sns.regplot(x=inc5000["revenue"], y=inc5000["growth"], fit_reg=False)
ax.set(xlabel='Revenue', ylabel='Growth %')
plt.show()
```
