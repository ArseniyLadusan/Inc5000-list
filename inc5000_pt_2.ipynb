```python
# NumPy for numerical computing
import numpy as np

# Pandas for DataFrames
import pandas as pd


# Matplotlib for visualization
from matplotlib import pyplot as plt
# display plots in the notebook
%matplotlib inline 

# Seaborn for easier visualization
import seaborn as sns

```

```python
inc5000 = pd.read_csv('inc5000_10years.csv', encoding = "ISO-8859-1")
```

```python
#Industry Countplot
sns.countplot(y='industry', data=inc5000[inc5000.metro == 'New York City']).set_title('New York City')
plt.show()

sns.countplot(y='industry', data=inc5000[inc5000.metro == 'Washington, DC']).set_title('Washington, DC')
plt.show()

sns.countplot(y='industry', data=inc5000[inc5000.metro == 'Philadelphia']).set_title('Philadelphia')
plt.show()
```

```python
#Facetgrid for NYC
incb = sns.FacetGrid(inc5000[inc5000.metro == 'New York City'], col="industry",col_wrap=5, hue="industry")
incb= incb.map(plt.hist, "year",)
plt.subplots_adjust(top=0.9,)
incb.fig.suptitle('New York City',fontsize=16)
incb.set_titles("{col_name}")
plt.show()
```

```python
#Employment graph by city
plt.rcParams['figure.figsize']=15,5
g= sns.stripplot(y='workers', x='industry', data=inc5000[inc5000.metro == 'New York City'],jitter=True)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title("New York City")
plt.show()
```
